```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# uploading my dataset

df = pd.read_csv('/kaggle/input/students-score-dataset-linear-regression/student_scores.csv')
df
```

**Data Visualization**

```python
plt.style.use('ggplot')
plt.title("Hours vs Scores")
plt.xlabel('Hours')
plt.ylabel('Score')
plt.scatter(df['Hours'],df['Scores'])
plt.plot()
```

**Data Set Preparing**

```python
X = df['Hours']
Y = df['Scores']
```

```python
from sklearn.model_selection import train_test_split

# data spliting into train and test

X_train,X_test,Y_train,Y_test  = train_test_split(X,Y,test_size =0.2, random_state = 10)
```

```python
len(X_train)
len(X_test)
len(Y_train)
len(Y_test)
```

**Model Training**

```python
reg_model = linear_model.LinearRegression()
```

```python
X_train = np.array(X_train).reshape(-1, 1)
reg_model.fit(X_train, Y_train)
```

```python
X_test = np.array(X_test).reshape(-1,1)
test_result = reg_model.predict(X_test)
```

**Model Result**

```python
X_test = np.array(X_test).reshape(-1)
df_result = pd.DataFrame({'X_test':X_test,'Y_test':Y_test,'Y-predicted':test_result})
df_result 
```

```python
# error finding

df_result['diff(yi - yp)'] = df_result['Y_test'] - df_result['Y-predicted']
df_result
```

**Perfroming Evaluation Matrix for Regression Model**

```python
X_test = np.array(X_test).reshape(-1,1)
reg_model.score(X_test,Y_test)
```

**R-squared**

```python
from sklearn.metrics import r2_score

r2_score = r2_score(X_test,Y_test)
r2_score
```

**Mean Squared Error (MSE)**

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(X_test,Y_test)
print(mse)
```

**Mean Absolute Error (MAE)**

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(X_test,Y_test)
mae
```
